```python
def get_qc_iteration(qci,tol=1e-8):
    qc=qci*1.0
    n=0
    x1=0.78
    x2=-0.33
    y1=-0.32
    y2=-0.35
    y3=0.49
    z1=1.21
    f1=x1*(qc**x2)
    f2=-(y1*(qc**y2)+y3)
    f3=abs((log10(10+qc))**z1)
    c=f1*((2.64/f3)**f2)
    cq=(101.0/771.258)**c
    while abs(qc-cq*qc)>tol:
        n+=1
        f1=x1*(qc**x2)
        f2=-(y1*(qc**y2)+y3)
        f3=abs((log10(10+qc))**z1)
        c=f1*((2.64/f3)**f2)
        cq=(101/771.258)**c
        qc=cq*qc
        #print ("f1=",f1, "f2=",f2, "f3=",f3, "c=",c, "cq=",cq)
        #print("%d times,qc = %f" %(n,qc))
    
        if n >100:
            break
    print("value of qc is:%f" % qc)
```

```python
get_qc_iteration(3.036)
```
