```python
A=1
u_list_f_all=[]
u_list_max_all=[]
av_list_all=[]
f_list=np.linspace(1,10,100)
for i,f in enumerate (f_list):
    w=2*np.pi*f
    t = np.linspace(0, 1/f, 101)
    a=A*np.sin(w*t)
    v = integrate.cumtrapz(a,t,initial=0)
    u= integrate.cumtrapz(v,t,initial=0)
    av=u[-1]*f
    u_list_f_all.append(u[-1])
    u_list_max_all.append(max(u))
    av_list_all.append(av)
```

```python
plt.figure(figsize=(9,6))
plt.subplot(1,2,1)
plt.plot(f_list, u_list_f_all,'g-*')
plt.subplot(1,2,2)
plt.plot(f_list, u_list_max_all,'r-o')
plt.xlabel('f')
plt.ylabel('final u')
plt.title('relationship between f and final u')
```

```python
def func_av_f(f,a,b):
    return a*f-b*np.sin(1/f)
```

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit(func_av_f, f_list, av_list_all)
#Optimal values for the parameters so that the sum of the squared residuals 
# of f(xdata, *popt) - ydata is minimized
```

```python
print (popt)
```

```python
plt.figure(figsize=(9,6))
plt.scatter(f_list,-0.0002884*f_list+0.17275358*np.sin(1/f_list),color="green")
plt.plot(f_list,func_av_f(f_list, *popt),color="red",label="fitting_curve: a=%5.7f, b=%5.7f" % tuple(popt))
plt.scatter(f_list,av_list_all,label="relation_av_to_f")
plt.xlabel('f')
plt.ylabel('av')
plt.legend()
plt.show()
```

```python
print ("av_f_fomular: av= -0.0002884*f+0.17275358*np.sin(1/f)")
```

```python
residuals = av_list_all- func_av_f(f_list, *popt)
ss_res = np.sum(residuals**2)
ss_tot = np.sum((av_list_all-np.mean(av_list_all))**2)
r_squared = 1 - (ss_res / ss_tot)
print (r_squared, "av_f_fitting_curve_R2")
```

```python
f=1
u_list_A=[]
av_list_A=[]
plt.figure(figsize=(9.8,15), dpi=100)
for j,A in enumerate ([0.2,0.4,0.6,0.8,1]):
    w=2*np.pi*f
    t = np.linspace(0, 1/f, 101)
    a=A*np.sin(w*t)
    v = integrate.cumtrapz(a,t,initial=0)
    u= integrate.cumtrapz(v,t,initial=0)
    u_list_A.append(u[-1])
    av=u[-1]*f
    av_list_A.append(av)
    print ("last u =",u[-1],"when f=",f)
    plt.subplot(3,2,j+1)
    plt.plot(t,a, 'b*-', label=u"a (m/s2)")
    plt.plot(t,v, 'ro-', label=u"v (m/s)")
    plt.plot(t,u, 'g--', label=u"u (m)")
    plt.xlabel('t (s)')
    plt.ylabel('m')
    plt.legend(['a', 'v', 'u'])
    plt.title("A="+ str(A))
```
